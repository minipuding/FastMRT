```python
"""显示热力图"""
import matplotlib.pyplot as plt
import numpy as np
import colorsys

data_type = "std"

if data_type == "terr":
    data = np.array(
        [[0.503,	0.447,	0.485,	0.487],
        [0.447,	0.454,	0.457,	0.457],
        [0.485,	0.457,	0.494,	0.493],
        [0.487,	0.457,	0.493,	0.495]])
    title = f"Temperature Error ({r'$T_{err}$'}, ℃)"
    cmap = "Reds_r"
elif data_type == "dice":
    data = np.array([
        [0.323,	0.398,	0.348,	0.371],
        [0.398,	0.403,	0.406,	0.396],
        [0.348,	0.406,	0.338,	0.332],
        [0.371,	0.396,	0.332,	0.334]
    ])
    title = "Patch Dice Coefficient (DICE)"
    cmap = "Reds"
elif data_type == "std":
    data = np.array([
        [1.448,	1.343,	1.407,	1.396],
        [1.343,	1.346,	1.354,	1.352],
        [1.407,	1.354,	1.428,	1.441],
        [1.396,	1.352,	1.441,	1.438]
    ])
    title = "Patch Standard Deviation (STD, ℃)"
    cmap = "Reds_r"
elif data_type == "rmse":
    data = np.array([
        [1.529,	1.388,	1.467,	1.451],
        [1.388,	1.386,	1.395,	1.394],
        [1.467,	1.395,	1.494,	1.505],
        [1.451,	1.394,	1.505,	1.497]
    ])
    title = "'Patch Root Mean Square Error (RMSE, ℃)"
    cmap = "Reds_r"

# 绘制热力图
fig, ax = plt.subplots()
im = ax.imshow(data, cmap=cmap)

# 添加颜色条
cbar = ax.figure.colorbar(im, ax=ax)

# 显示数值
for i in range(4):
    for j in range(4):
        # 获取背景颜色的亮度值
        r, g, b, _ = im.cmap(im.norm(data[i, j]))
        h,s, v = colorsys.rgb_to_hsv(r, g, b)

        # 根据背景颜色的亮度值选择数字颜色
        if v > 0.5:
            text_color = 'black'
        else:
            text_color = 'white'

        # 在方格中添加数字
        text = ax.text(j, i, f'{data[i, j]:.3f}',
                       ha='center', va='center', color=text_color, fontsize=18, fontname='Times New Roman')

# 设置坐标轴标签和标题
ax.set_xticks(np.arange(4))
ax.set_yticks(np.arange(4))
ax.set_xticklabels(['DA', 'CA', 'KD', 'DL'], fontsize=18, fontname='Times New Roman')
ax.set_yticklabels(['DA', 'CA', 'KD', 'DL'], fontsize=18, fontname='Times New Roman')
ax.set_title(title, fontsize=18, fontname='Times New Roman')

# 调整图像大小和布局
fig.tight_layout()

# 显示图像
plt.show()
```
